9. Format 2D array
def format_2d_array(arr):
    width = max(len(str(x)) for row in arr for x in row)
    for row in arr:
        print(" ".join(f"{x:>{width}}" for x in row))

15. Median & Mode
def median(arr):
    return np.median(arr)

def mode(arr):
    vals, counts = np.unique(arr, return_counts=True)
    return vals[counts.argmax()]
