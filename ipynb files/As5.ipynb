11. Number to words
def num_to_words(n):
    d = {
        '0':'Zero','1':'One','2':'Two','3':'Three','4':'Four',
        '5':'Five','6':'Six','7':'Seven','8':'Eight','9':'Nine'
    }
    return ' '.join(d[i] for i in str(n))

18. Analyze sets
def analyze_sets(list1, list2):
    s = set(list1) ^ set(list2)
    result = []
    for x in s:
        if x % 2 == 0:
            result.append(x*2)
        else:
            result.append(x+5)
    return sorted(result)

19. Unique word pairs
def unique_pairs(words):
    result = set()
    words = [w for w in words if len(w) >= 4]

    for i in range(len(words)):
        for j in range(i+1, len(words)):
            if set(words[i]).isdisjoint(set(words[j])):
                pair = tuple(sorted((words[i], words[j])))
                result.add(pair)
    return result
