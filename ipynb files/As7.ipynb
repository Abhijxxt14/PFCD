1. Replace successive repetitive characters with *
def replace_repetitive(s):
    result = s[0]
    for i in range(1, len(s)):
        if s[i] == s[i-1]:
            result += '*'
        else:
            result += s[i]
    return result

print(replace_repetitive("balloon"))  # bal*o*n

6. Replace vowels with next vowel
def replace_vowels(s):
    vowels = "aeiou"
    result = ""
    for c in s:
        if c in vowels:
            result += vowels[(vowels.index(c)+1) % 5]
        else:
            result += c
    return result

7. Rotational Palindrome
def is_rotational_palindrome(s):
    for i in range(len(s)):
        if (s[i:] + s[:i]) == (s[i:] + s[:i])[::-1]:
            return True
    return False
