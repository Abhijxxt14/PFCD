def can_form_palindrome(s):

13. Roman to Integer
def roman_to_int(s):
    r = {'I':1,'V':5,'X':10,'L':50,'C':100,'D':500,'M':1000}
    total = 0
    for i in range(len(s)):
        if i+1 < len(s) and r[s[i]] < r[s[i+1]]:
            total -= r[s[i]]
        else:
            total += r[s[i]]
    return total

14. Armstrong number
def is_armstrong(n):
    digits = list(map(int, str(n)))
    p = len(digits)
    return n == sum(d**p for d in digits)

15. Nth Fibonacci
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n-1):
        a, b = b, a+b
    return a

20. Prime numbers (1–100)
def primes_1_to_100():
    primes = []
    for n in range(2, 101):
        for i in range(2, n):
            if n % i == 0:
                break
        else:
            primes.append(n)
    return primes
