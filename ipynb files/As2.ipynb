7. Sum of primes below n
n = int(input("Enter number: "))
s = 0
for num in range(2, n):
    for i in range(2, int(num**0.5)+1):
        if num % i == 0:
            break
    else:
        s += num
print(s)

    if op == "+":
        print(a+b)
    elif op == "-":
