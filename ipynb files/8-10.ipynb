```python
#Q1 (a)

#Remove duplicate elements from a list while preserving original order

lst = list(map(int, input("Enter integers: ").split()))

result = []
seen = set()

for x in lst:
    if x not in seen:
        result.append(x)
        seen.add(x)

print("List after removing duplicates:", result)
```

```python
#Q2 (a)

#Average weekly temperature & count days above average (NumPy)

import numpy as np

temps = np.array(list(map(float, input("Enter 7 temperatures: ").split())))

avg = np.mean(temps)
count = np.sum(temps > avg)

print("Average temperature:", avg)
print("Days above average:", count)
```

```python
#Q2 (b)

#Total sales per product & product with highest sales (NumPy)

import numpy as np

rows = int(input("Enter number of products: "))
cols = int(input("Enter number of days: "))

sales = []

for i in range(rows):
    while True:
        data = list(map(int, input(f"Enter sales for product {i+1}: ").split()))
        if len(data) == cols:
            sales.append(data)
            break
        else:
            print("Please enter exactly", cols, "values")

sales = np.array(sales)

total_sales = np.sum(sales, axis=1)
highest_product = np.argmax(total_sales) + 1

print("Total sales of each product:", total_sales)
print("Product with highest sales: Product", highest_product)
```
